```python
import requests
import pandas as pd
# Import API key
from api import API_key
```

```python
url = "https://api.yelp.com/v3/businesses/search"
headers = {
        'Authorization': 'Bearer ' + API_key,
        'Content-Type': 'application/json'
    }
```

```python
restaurants= []
```

```python
def call_api(states, offsets):
    for state in states:
        print(f'Getting data for state: {state}')
        for offset in offsets:
            print(f'Getting offset: {offset}')
            params = {'location': state,
                'limit': 50,
                'offset': offset,
                'sort_by': 'review_count',
                'term': 'food'
                }
            response = requests.get(url,  headers=headers, params=params).json()
            if 'businesses' in response:
                restaurants.extend(response['businesses'])
            
```

```python
states = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida", "Georgia"]
states += ["Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine", "Maryland"]
states += ["Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey"]
states += ["New Mexico", "New York", "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina"]
states += ["South Dakota", "Tennessee", "Texas", "Utah", "Vermont" ]
states += ["Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"]
offsets = [0,51,101,151,201,251,301,351,401,451]
# Calling extract api
call_api(states, offsets)
```

```python
df = pd.DataFrame(restaurants)
df.head(3)
```
